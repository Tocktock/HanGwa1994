```python
import json
import re
import nltk
from gensim import models
from gensim import corpora
import numpy as np;np.random.seed(42)
# -*- coding: utf-8 -*-
ntopics, nwords = 5,100
with open('Tr.json', encoding="iso-8859-1") as f:
    json_data = json.load(f, encoding="utf-8")
    regex = re.compile(r'[가-힣]+/NNG')
    noNNG = re.compile(r'[가-힣]+')
    reAns = re.compile(r'[0-9]+')
    x_train = []
    y_train = []
    for i in range(len(json_data)):
        TrainX = []
        TrainY = []
        for y in json_data[i]['answer']:
            y = reAns.findall(y)
            TrainY = TrainY + y
        #print(TrainY)
        y_train.append(TrainY)
        for j in json_data[i]['sentence']:
            sentence = j.encode('iso-8859-1').decode('euc-kr', 'ignore')
            matchWord = regex.findall(sentence)
            sentence = ''.join(matchWord)
            matchWord = noNNG.findall(sentence)
            TrainX = TrainX + matchWord
        x_train = x_train + TrainX
        
dataset = [d.split() for d in x_train]
dicko = corpora.Dictionary(dataset)
tf_ko = [dicko.doc2bow(text) for text in dataset]
tfidfModel = models.TfidfModel(tf_ko)
tfidf_ko = tfidfModel[tf_ko]
#LSI
lsi_ko = models.lsimodel.LsiModel(tf_ko,id2word = dicko,num_topics = ntopics)
#LDA
lda_ko = models.ldamodel.LdaModel(tfidf_ko,id2word = dicko,num_topics = ntopics)
#print(lda_ko.print_topics(num_topics=ntopics, num_words=nwords))

bow = tfidfModel[dicko.doc2bow(dataset[0])]
sorted(lsi_ko[bow], key=lambda x: x[1], reverse=True)
sorted(lda_ko[bow], key=lambda x: x[1], reverse=True)
        

```
